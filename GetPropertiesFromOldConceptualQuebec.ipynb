```python
import xml.etree.ElementTree as ET
```

```python
NAME = 'Name'
AREA = 'Area'

N_PEOPLE = 'NumberPeople'
TIMEPATTERN ='TimePattern'
DFLOW_BASELINE = 'Baseline'

VMAX = 'Vmax'
K = 'k'
TANK_INDEXES = 'tanksIndexes'
LEN = 'Length'

END = 'EndNode'
```

```python
XML_SEWER_NAMES =  'Sew_'
XML_CATCH_NAMES = 'Catchment_'
XML_CONN_NAMES = 'Connector_info_'

XML_AREA_SEWER = '.A'
XML_VMAX_SEWER = '.V_Max'
XML_K_SEWER = '.k'

XML_AREA_CATCH = '.TotalArea'

XML_CONN_VELCLASS = '.Vs_limit'
XML_CONN_VELMINCLASS = '.Vs_limit_min'

XML_DWF_NPEOPLE = '.Inhabitants' #Associated to a catchment in WEST
XML_DWF_CUSTOMFLOWPATTERN = '.DWF.CustomFlow(H' #Associated to a catchment in WEST
XML_DWF_INFILTRATION = '.DWF.Infiltration' #Associated to a catchment in WEST

XML_MODEL_PROP_NAME = 'InstanceDisplayName'


XML_QUANTIY = 'Quantity'
```

```python
CATCH_PROPER = [XML_AREA_CATCH,XML_DWF_NPEOPLE]
DWF_PATTERN = [f"{XML_DWF_CUSTOMFLOWPATTERN}{hour:02})" for hour in range(1, 24)]
CATCH_PROPER.extend(DWF_PATTERN)

SEWER_PROPER = [XML_AREA_SEWER,XML_VMAX_SEWER,XML_K_SEWER,LEN]


CATCH_DICT = [AREA,N_PEOPLE]
DWF_PAT_DICT = [TIMEPATTERN for i in range(1,24)]
CATCH_DICT.extend(DWF_PAT_DICT)


SEWER_DICT = [AREA,VMAX,K]
```

```python
def getWESTProperties(xmlFile,catchments,sewers):

    tree = ET.parse(xmlFile)
    root = tree.getroot()  
    quantities_elem = root.find('.//Quantities')

    listCatchments = []
    listSewers = []
    
    for elements,properties,pEqu,listE in zip([catchments,sewers],[CATCH_PROPER,SEWER_PROPER],[CATCH_DICT,SEWER_DICT],[listCatchments,listSewers]):

        for element in elements:
        
            dictionaryEle = {}
            dictionaryEle[NAME] = element
            listPattern = []
            
            for p, pInDic in zip(properties, pEqu):
            
                valQuantityName = '.' + element + p
                valQuantity = root.find(".//Quantities/Quantity[@Name='"+ valQuantityName +"']")
    
                if (p in DWF_PATTERN):

                    if (valQuantity is not None):
                    
                        val= valQuantity.find("./Props/Prop[@Name='Value']").get('Value')
                        listPattern.append(val)
                    else:
                        listPattern.append(val)
                
                else (valQuantity is not None):
                    
                    val= valQuantity.find("./Props/Prop[@Name='Value']").get('Value')
                    dictionaryEle[pInDic] = val
                    
                else:
                    print(valQuantityName, " was not found")

            
            if listPattern:
                dictionaryEle[TIMEPATTERN] = listPattern

            listE.append(dictionaryEle)

    return listCatchments,listSewers
```

```python
xmlFile = '01-Data/01-OldWESTModel/Catchment_Fault.Layout.xml'


catchments = [XML_CATCH_NAMES + str(i) for i in range(1, 12)]
sewers = [XML_SEWER_NAMES + str(i) for i in range(1, 93)]



cat, sew = getWESTProperties(xmlFile,catchments,sewers)


for c in cat:
    c[DFLOW_BASELINE] = None
    c[END] = True

i =0
for s in sew:
    i += 1 
    s[TANK_INDEXES] = [i]
    s[LEN]=99
```
